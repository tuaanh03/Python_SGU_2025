Chủ đề 4 – Thêm cột mới

Ví dụ 1: Thêm cột đánh giá (Pass/Fail)

```python
df["result"] = df["math_score"].apply(lambda x: "Pass" if x >= 75 else "Fail")
df.head()
```

Ví dụ 2: Thêm cột nhóm tuổi

```python
def age_group(age):
    if age <= 19:
        return "Teen"
    elif age <= 21:
        return "Young"
    return "Adult"

df["age_group"] = df["age"].apply(age_group)
df.head()
```

4. 10 chủ đề nhỏ với NumPy (mỗi chủ đề 2 ví dụ)

```python
import numpy as np
import pandas as pd
```

Đọc dữ liệu

```python

df = pd.read_csv("students.csv")   # hoặc students.xlsx
df.head()
```

```python
ages = df["age"].to_numpy()
math_scores = df["math_score"].to_numpy()
```

🎯 CHỦ ĐỀ 1 – Chuyển DataFrame → NumPy & xem thông tin

Ví dụ 1: Lấy mảng điểm toán

```python
math_scores = df["math_score"].to_numpy()
print(math_scores)
print(math_scores.dtype, math_scores.shape)
```

Ví dụ 2: Lấy toàn bộ numeric columns

```python
numeric_df = df[["age", "math_score"]]
numeric_matrix = numeric_df.to_numpy()
print(numeric_matrix)
print(numeric_matrix.shape)

```

CHỦ ĐỀ 2 – Thống kê cơ bản với NumPy

Ví dụ 1: Tính mean, median, std điểm toán

```python
print("Mean:", math_scores.mean())
print("Median:", np.median(math_scores))
print("Std:", math_scores.std())

```

Ví dụ 2: Tìm min / max và vị trí của nó

```python
print("Min score:", math_scores.min())
print("Max score:", math_scores.max())
print("Index of max:", math_scores.argmax())
```

CHỦ ĐỀ 3 – Chuẩn hóa dữ liệu (Normalization)

Ví dụ 1: Min-Max Scaling

```python
min_val = math_scores.min()
max_val = math_scores.max()
scaled = (math_scores - min_val) / (max_val - min_val)
print(scaled)
```

Ví dụ 2: Scaling với Z-Score

```python
z = (math_scores - math_scores.mean()) / math_scores.std()
print(z)
```

CHỦ ĐỀ 4 – Broadcasting

(cho phép thực hiện các phép toán trên mảng có hình dạng khác nhau)

Ví dụ 1: Cộng toàn bộ điểm toán thêm 5

```python
bonus = math_scores + 5
print(bonus)
```

Ví dụ 2: Trừ tuổi mỗi bạn đi tuổi trung bình

```python
age_centered = ages - ages.mean()
print(age_centered)
```

CHỦ ĐỀ 5 – Boolean Mask

Ví dụ 1: Lọc điểm > 80

```python
boolean_mask = df["math_score"] > 80
filtered_data = df[boolean_mask]
print(filtered_data)
```

Ví dụ 2: Lấy tuổi của học sinh ở HCM (kết hợp DF → NP)

```python
mask = df["city"] == "HCM"
ages_hcm = ages[mask.to_numpy()]
print(ages_hcm)
```

CHỦ ĐỀ 6 – Xử lý missing value bằng NumPy

Giả lập trường hợp bị NaN

```python
scores_nan = math_scores.astype("float")
scores_nan[3] = np.nan
scores_nan[7] = np.nan
```

Vd1: Tìm vị trí bị Null

```python
nan_index = np.where(np.isnan(scores_nan))
print(nan_index)
```

Ví dụ 2: Điền giá trị trung bình vào vị trí bị null

```python
mean_val = np.nanmean(scores_nan)
scores_filled = np.where(np.isnan(scores_nan), mean_val, scores_nan)
print(scores_filled)
```

CHỦ ĐỀ 8 – Random với dữ liệu thật

Ví dụ 1: Lấy ngẫu nhiên 5 học sinh

```python
idx = np.random.choice(len(df), 5, replace=False)
random_students = df.iloc[idx]
print(random_students)
```

Ví dụ 2: Tạo độ nhiễu (noise) cho điểm toán (mô phỏng 1 dữ liệu không hoàn hảo)

```python
noise = np.random.randint(-3, 4, size=len(math_scores))
noisy_scores = math_scores + noise
print(noisy_scores)
```

CHỦ ĐỀ 9 – Tính correlation (tương quan)

Ví dụ 1: Tương quan giữa age và math_score

```python
corr = np.corrcoef(ages, math_scores)
print(corr)
```

Ví dụ 2: Tương quan nhiều cột

```python
features = df[["age", "math_score"]].to_numpy().T
corr_all = np.corrcoef(features)
print(corr_all)
```

CHỦ ĐỀ 10 – Vectorized operations nâng cao

Ví dụ 1: Chuẩn hóa toàn bộ numeric matrix (ages + math_score)

```python
X = df[["age", "math_score"]].to_numpy()

X_norm = (X - X.mean(axis=0)) / X.std(axis=0)
print(X_norm)
```

Ví dụ 2: Tính khoảng cách Euclidean giữa 2 học sinh

```python
v1 = X[0]
v2 = X[1]
dist = np.linalg.norm(v1 - v2)
print(dist)
```
